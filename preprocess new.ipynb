Preprocessing - Image Resizing and Normalization

```python
import os
import cv2
import numpy as np

# Define path to dataset
dataset_path = "/content/drive/MyDrive/Dataset/adenosis"

# Define path to save preprocessed images
save_path = "/content/drive/MyDrive/Preprocessed_data/adenosis"

# Define image size to resize images to
img_size = (256, 256)

# Loop through all subdirectories in the dataset
for subdir in os.listdir(dataset_path):
    if os.path.isdir(os.path.join(dataset_path, subdir)):
        print("Processing images in directory:", subdir)
        for filename in os.listdir(os.path.join(dataset_path, subdir)):
            if filename.endswith(".png"):
                # Read image
                img = cv2.imread(os.path.join(dataset_path, subdir, filename))

                # Resize image
                img_resized = cv2.resize(img, img_size)

                # Normalize image
                img_normalized = img_resized / 1

                # Save preprocessed image
                save_filename = os.path.join(save_path, subdir, filename)
                os.makedirs(os.path.dirname(save_filename), exist_ok=True)
                cv2.imwrite(save_filename, img_normalized)

                print("Processed image:", save_filename)
```

GAN

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing import image

# Define path to dataset
dataset_path = "/content/drive/MyDrive/Preprocessed_data"

# Define path to save generated images
save_path = "/content/drive/MyDrive/Preprocessed_data/GAN"

# Define input shape of generator
generator_input_shape = (100,)

# Define number of epochs and batch size
epochs = 10000
batch_size = 128

# Load and preprocess dataset
def load_dataset():
    X_train = []
    for subdir in os.listdir(dataset_path):
        if os.path.isdir(os.path.join(dataset_path, subdir)):
            for filename in os.listdir(os.path.join(dataset_path, subdir)):
                if filename.endswith(".png"):
                    img = image.load_img(os.path.join(dataset_path, subdir, filename), target_size=(28, 28))
                    x = image.img_to_array(img)
                    x = (x - 127.5) / 127.5
                    X_train.append(x)
    X_train = np.array(X_train)
    return X_train

# Define generator
def build_generator(input_shape):
    model = Sequential()
    model.add(layers.Dense(128 * 7 * 7, activation="relu", input_dim=input_shape))
    model.add(layers.Reshape((7, 7, 128)))
    model.add(layers.UpSampling2D())
    model.add(layers.Conv2D(64, kernel_size=3, padding="same", activation="relu"))
    model.add(layers.UpSampling2D())
    model.add(layers.Conv2D(1, kernel_size=3, padding="same", activation="tanh"))
    return model

# Define discriminator
def build_discriminator(input_shape):
    model = Sequential()
    model.add(layers.Conv2D(64, kernel_size=3, strides=2, input_shape=input_shape, padding="same"))
    model.add(layers.LeakyReLU(alpha=0.2))
    model.add(layers.Dropout(0.4))
    model.add(layers.Conv2D(128, kernel_size=3, strides=2, padding="same"))
    model.add(layers.LeakyReLU(alpha=0.2))
    model.add(layers.Dropout(0.4))
    model.add(layers.Flatten())
    model.add(layers.Dense(1, activation="sigmoid"))
    return model

# Combine generator and discriminator
def build_gan(generator, discriminator):
    discriminator.trainable = False
    model = Sequential()
    model.add(generator)
    model.add(discriminator)
    model.compile(loss="binary_crossentropy", optimizer=Adam(lr=0.0002, beta_1=0.5))
    return model

# Train GAN
def train_gan(X_train, generator, discriminator, gan, epochs, batch_size):
    for epoch in range(epochs):
        # Train discriminator
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_images = X_train[idx]
        noise = np.random.normal(0, 1, (batch_size, generator_input_shape))
        generated_images = generator.predict(noise)
        X = np.concatenate((real_images, generated_images))
        y = np.concatenate((np.ones((batch_size, 1)), np.zeros((batch_size, 1))))
        d_loss = discriminator.train_on_batch(X, y)

        # Train generator
            noise = np.random.normal(0, 1, (batch_size, generator_input_shape))
    y = np.ones((batch_size, 1))
    g_loss = gan.train_on_batch(noise, y)

    # Print loss every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch: {epoch}, Discriminator Loss: {d_loss}, Generator Loss: {g_loss}")
    
    # Save generated images every 500 epochs
    if epoch % 500 == 0:
        noise = np.random.normal(0, 1, (25, generator_input_shape))
        generated_images = generator.predict(noise)
        generated_images = 0.5 * generated_images + 0.5
        fig, axs = plt.subplots(5, 5)
        count = 0
        for i in range(5):
            for j in range(5):
                axs[i, j].imshow(generated_images[count, :, :, 0], cmap="gray")
                axs[i, j].axis("off")
                count += 1
        fig.savefig(os.path.join(save_path, f"generated_images_epoch{epoch}.png"))
        plt.close()
```
